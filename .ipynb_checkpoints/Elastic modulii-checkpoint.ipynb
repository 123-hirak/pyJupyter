```python
import numpy as np
import matplotlib.pyplot as plt

# Elastic moduli calculation function
def calc_moduli(vp, vs,rho):
    # Young's modulus
    E = rho * vp**2 * (3 * vs**2 + 4 * vp**2) / (vp**2 - 4 * vs**2)
    # Shear modulus
    G = rho * vs**2
    # Bulk modulus
    K = rho * vp**2 - 4/3 * G
    # Poisson ratio
    nu = (vp**2 - 2 * vs**2) / (2 * (vp**2 - vs**2))
    return E, G, K, nu

# Sample data
vp = 5500 # m/s
vs = 2300 # m/s
rho =2710 # kg/m3


# Calculate moduli for the original data
E,G,K,nu = calc_moduli(vp, vs, rho)

# Errors in the measurements
err_vp = 5 # m/s
err_vs = 6 # m/s



# Calculate moduli with errors in vp and vs
E_err_vp = calc_moduli(vp + err_vp, vs, rho)[0]
E_err_vs = calc_moduli(vp, vs + err_vs, rho)[0]


# Calculate percent error in Young's modulus
pct_err_vp = abs((E_err_vp - E) / E) * 100
pct_err_vs = abs((E_err_vs - E) / E) * 100
print(f"Percent error with 1 m/s error in P-wave: {pct_err_vp:.2f}%")
print(f"Percent error with 1 m/s error in S-wave: {pct_err_vs:.2f}%")

# Plot results
x_labels = ['Original',  '1 m/s error in P-wave',   '1 m/s error in S-wave']
y_values = [E, E_err_vp, E_err_vs]
pct_errors = [0, pct_err_vp, pct_err_vs]

fig, ax1 = plt.subplots()
ax1.bar(x_labels, y_values, color='C0')
ax1.set_ylabel('Young\'s modulus (GPa)')
ax2 = ax1.twinx()
ax2.plot(x_labels, pct_errors, color='C1', marker='o', markersize=8)
ax2.set_ylabel('Percent error in Young\'s modulus')
ax2.set_ylim(0, 10)
plt.title('Effect of measurement errors on Young\'s Modulus')

plt.show()
#fig.savefig("Youngs_modulus.jpeg")
```
