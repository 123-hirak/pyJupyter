```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
time = np.arange(0, 10, 0.1)  # 100 time steps from 0 to 10
amplitude = np.sin(time) + 0.5 * np.sin(3 * time)  # a combination of 2 sine waves

# Compute the FFT of the amplitude data
fft_amplitude = np.fft.fft(amplitude)

# Calculate the frequency domain of the FFT data
n = len(amplitude)
freq = np.fft.fftfreq(n, d=(time[1]-time[0]))

# Plot the time domain data
fig, axs = plt.subplots(2, 1, figsize=(10, 6))
axs[0].plot(time, amplitude)
axs[0].set_xlabel('Time')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Time Domain')

# Plot the frequency domain data
axs[1].plot(freq, np.abs(fft_amplitude))
axs[1].set_xlabel('Frequency (Hz)')
axs[1].set_ylabel('Power')
axs[1].set_xlim(0, 5)  # show only frequencies up to 5 Hz
axs[1].set_title('Frequency Domain')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# generate a signal
t = np.linspace(0, 1, 1000)
x = np.sin(2 * np.pi * 5 * t) + np.sin(2 * np.pi * 10 * t)

# compute the FFT
X = np.fft.fft(x)

# calculate the frequency bins
freq = np.fft.fftfreq(len(x), t[1] - t[0])

# plot the signal and FFT
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.plot(t, x)
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax2.plot(freq, np.abs(X))
ax2.set_xlabel('Frequency')
ax2.set_ylabel('Magnitude')
plt.show()
```
