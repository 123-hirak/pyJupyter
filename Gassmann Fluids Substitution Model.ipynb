```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load Excel file
df = pd.read_excel('F:/Gassman.xlsx')
df.head()
```

```python
import plotly as py
import plotly.graph_objs as go
# Extract Measured Vp and Calculated Vp for brine and air
measured_vp = df['Vp(km/s)']
calculated_vp_brine = df['Vp_Cal(brine)']
calculated_vp_air = df['Vp_Cal(air)']
calculated_vp_oil = df['Vp_Cal(oil)']
calculated_vp_CO2 = df['Vp_Cal(CO2)']

# Plot data
plt.scatter(measured_vp, calculated_vp_brine, label='Calculated Vp (Brine)',marker ='o',color = 'red',s=60)
plt.scatter(measured_vp, calculated_vp_air, label='Calculated Vp (Air)',marker ='*',color ='yellow')
plt.scatter(measured_vp, calculated_vp_oil, label='Calculated Vp (Oil)',marker ='^',color ='blue')
plt.scatter(measured_vp, calculated_vp_CO2, label='Calculated Vp (CO2)',marker ='+',color ='purple')
plt.xlabel('Measured Compressional Wave Velocity(km/s)')
plt.ylabel('Calculated Compressional Wave Velocity(km/s)')
plt.title ('Gassmann\'s Fluids Substitution For Compressional Wave Velocity')
plt.legend()

# Draw intercept line
fit = np.polyfit(measured_vp, calculated_vp_air, 1)
intercept = fit[1]
slope = fit[0]
x_range = np.linspace(min(measured_vp), max(measured_vp), 100)
y_range = slope * x_range + intercept
plt.plot(x_range, y_range, color='black',alpha =0.5)
plt.savefig('Gassman(Vp).jpeg',dpi=1500)
plt.show()
```

```python
import plotly as py
import plotly.graph_objs as go
# Extract Measured Vp and Calculated Vp for brine and air
measured_vs = df['Vs1(km/s)']
calculated_vs_brine = df['Vs_Cal(brine)']
calculated_vs_air = df['Vs_Cal(air)']
calculated_vs_oil = df['Vs_Cal(oil)']
calculated_vs_CO2 = df['Vs_Cal(CO2)']

# Plot data
plt.scatter(measured_vs, calculated_vs_brine, label='Calculated Vs (Brine)',marker ='o',color = 'red',s=60)
plt.scatter(measured_vs, calculated_vs_air, label='Calculated Vs (Air)',marker ='*',color ='yellow')
plt.scatter(measured_vs, calculated_vs_oil, label='Calculated Vs (Oil)',marker ='^',color ='blue')
plt.scatter(measured_vs, calculated_vs_CO2, label='Calculated Vs (CO2)',marker ='+',color ='purple')
plt.xlabel('Measured Shear Wave Velocity(km/s)')
plt.ylabel('Calculated Shear Wave Velocity(km/s)')
plt.title ('Gassmann\'s Fluids Substitution For Shear Wave Velocity')
plt.legend()

# Draw intercept line
fit = np.polyfit(measured_vs, calculated_vs_air, 1)
intercept = fit[1]
slope = fit[0]
x_range = np.linspace(min(measured_vs), max(measured_vs), 100)
y_range = slope * x_range + intercept
#plt.xlim(0, 5)
#plt.ylim(0, 5)
plt.plot(x_range, y_range, color='black',alpha=0.5)
plt.savefig('Gassman(Vs).jpeg',dpi=1500)
plt.show()
```
